```python
df_country = pd.read_excel('E:\sel\chrome_test_profile\Zomatodataset\Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
## To check data types
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## pie chart. Top 3 countries that uses zomato
plt.pie(country_val[:3],labels = country_names[:3], autopct='%1.0f%%')
```

Observation:
Zomato has the biggest base in India followed by US & UK. Its maximum records or transactions are from India.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 ---> Excellent
2. When rating is between 4.0 to 4.5 ---> Very Good
3. When rating is between 3.5 to 3.9 ---> Good
4. 2148 users have not rated the food.

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)

sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count',hue = 'Rating color', data = ratings, palette=['blue','red','orange','yellow','green','green'])
```

Observation:
1. Not Rated Count is very high
2. Max no. of ratings are between 2.5 to 3.4
    

```python
## count plot
sns.countplot(x='Rating color',data=ratings, palette = ['blue','red','orange','yellow','green','green'])
```
